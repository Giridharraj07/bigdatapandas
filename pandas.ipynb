```python
# merging two dataframes
data1 =  {  'Name' : ['Jai', 'Anuj', 'Jai', 'Prince'],
           'Age' : [27 ,24 ,22, 32],
          'Address': ['Nagpur' , 'Kanpur' , 'Allahabad' , 'Kannauj'],
           'Qualification' : ['Msc', 'Ma', 'Mca','Phd'],
            'Salary' : [3423 ,43234 ,34243 , 546556 ]
        }
data2 =  {  'Name' : [ 'Gaurav' ,'Anuj' ,'Prince' , 'Abhi','Gaurav'],
           'Age' : [ 33, 36, 27, 32, 28],
          'Address': [ 'Jaunpur' , 'Kanpur' , 'Allahabad', 'Aligarh',' Varanasi'],
           'Qualification' : ['B.Tech' , 'B.com' , 'Msc' , 'Ma', 'B.Tech'],
            'Salary' : [ 64633 , 653565 , 645345 , 6456466 , 7564564]
        }
```

```python
#converting dict to df 
#manually indexing important to numberwise indexing other wise indexing geneated system wise example df1-0,1,2 and df2-0,1,2
# not like df1 0,1,2 and df2 - 3,4,5
df1 = pd.DataFrame(data1, index=[0,1,2,3])
df2 = pd.DataFrame(data2, index=[4,5,6,7,8])
```

```python
df1
```

```python
df2
```

```python
#merging the data frames
df_concat = [df1,df2]
res = pd.concat(df_concat)
res
```
